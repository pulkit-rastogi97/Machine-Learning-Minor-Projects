```python
#importing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading data
d=pd.read_csv('calcium.csv')
x=d.iloc[:,0]
y=d.iloc[:,1]
plt.plot(x,y,"r")
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
n=len(x)
numerator=0
denominator=0
for i in range(n):
    numerator=numerator+((x[i]-x_mean)*(y[i]-y_mean))
    denominator=denominator+((x[i]-x_mean)**2)
m=numerator/denominator
c=y_mean-(m*x_mean)
```

```python
min_x=np.min(x)
max_x=np.max(x)
min_y=np.min(y)
max_y=np.max(y)
x1=np.linspace(min_x,max_x)
y1=c+m*x1
plt.plot(x1,y1,"r")
```
